##  PerformanceMedia

+ When lvs_category = PerformanceMedia ï¼Œthere are 'Display', 'PaidSearch', 'Partner' three channel
+ Look at every channel, research the relationship between spend, leads, and visits.
+ The results shows that the relationship between spend and leads is strong, but spend has nothing to do with visit, there is some positive correlation between leads and visits
+ We think that the spending on these channel maybe not worthwhile, input are unrelated to outputs 

````python
condition=(df_lvs["lvs_category"]=="PerformanceMedia")

df_tmp=df_lvs[condition].copy()
df_tmp["lvs_detail"]=df_tmp["lvs_detail"].replace("Partners","Partner")

lst=['Display', 'PaidSearch', 'Partner']
for l in lst:
    df_tmp2=df_tmp[df_tmp["lvs_detail"]==l].copy()

    df_weekly_lvs_pm_pivot_tmp=df_tmp2.pivot_table(index="week_start",columns=["lvs"],values="value").fillna(0)
    for col in ['Leads', 'Spends', 'Visits']:
        # log transformation  The original numbers are large, take the log transformation, remove some outliers
        df_weekly_lvs_pm_pivot_tmp[col]=np.log1p(df_weekly_lvs_pm_pivot_tmp[col])
    
#     plt.figure(figsize=(5,5))
#     sns.pairplot(df_weekly_lvs_pm_pivot_tmp,corner=True)
#     plt.title("Pair Plot of {}".format(col),fontsize=12)
    
    df_tmp_corr=df_weekly_lvs_pm_pivot_tmp.corr()
    print("="*60)
    print("Correlation Coefficients ({})".format(l))
    display(df_tmp_corr)    
    # correlation 
    plt.figure(figsize=(8,4))
    sns.heatmap(df_tmp_corr, cmap="YlGnBu",annot=True, fmt="0.2f",linewidths=.5)
    plt.title("Correlation Heatmap ({})".format(l),fontsize=12)
    plt.show()
    
````

# model data selection

+ Based on previous category of data, we choose the partial data for model analysis later.
+ There is no total revenue for each channel. We can't calculate ROI of each channel directly. We can research the influence of each channel on the final revenue and other DV.  
+ In this part, modeling will focus on several channel's situation ï¼ˆspends visits leads). The impact of the variable on income or other goal Y. Other variables can be added.

+ Attentionï¼š Since three of the four different targets represent the nature of revenue and one represents the total visits, the x used in our model should not include revenue and visits fields 
    - "Total_DeseasonalizedMarketplace_Revenue","Total_Marketplace_SeasonalIndex","RevenueByRetail_Marketplace_Revenue"]
    - "Total_Total_Visits"
    

````python
from scipy.stats import pearsonr
````

````python

comp_cols=[x for x in df.columns if x.startswith("Competitor_") ]
ls_cols=[x for x in df.columns if (x.endswith("Spends"))]
macro_cols=[x for x in df.columns if x.startswith("Macro_") ]
inv_cols=[x for x in df.columns if x.endswith("_Inventory") ]
targets=["Total_DeseasonalizedMarketplace_Revenue","Total_Marketplace_SeasonalIndex","Total_Total_Visits","RevenueByRetail_Marketplace_Revenue"]
````

````python
df_model=df[comp_cols+ls_cols+macro_cols+inv_cols+targets].copy()
df_model.head()
````

## Missing values

+ The inventory field is missing a lot, for the real business data, we are not able to use any kinds of mean value to fill in. So we decided to delete them 
+ There are few other missing fields, let's use 0 to fill in the missing values

````python
df_missing=df.isnull().sum().reset_index()
df_missing.columns=["feature","missing count"]
df_missing["missing percentage"]=df_missing["missing count"]/df.shape[0]
df_missing["missing percentage"]=df_missing["missing percentage"].map(lambda x:"{:0.2%}".format(x))
df_missing.sort_values(by="missing count",ascending=False,inplace=True)

# unique values count
df_nunique=df.nunique().reset_index()
df_nunique.columns=["feature","unique values"]

# data types
df_types=df.dtypes.reset_index()
df_types.columns=["feature","dtype"]

# merge
df_stats=pd.merge(df_types,df_nunique,on="feature")
df_stats=pd.merge(df_stats,df_missing,on="feature")

# sort df
df_stats=df_stats.sort_values(by=["missing count"],ascending=[False])

df_stats.head(20)
````

## Compute the correlation

````python
# compute pearson correlation and p-value 

Xs=comp_cols+ls_cols+macro_cols

corr_list=[]
for x in Xs:
    for y in targets:
        
        corr,pvalue=pearsonr(df_model[x].fillna(0),df_model[y].fillna(0))
        corr_list.append({"x":x,"y":y,"corr":corr,"pvalue":pvalue})
        
df_corr=pd.DataFrame(corr_list)

# keep the p value < 0.2 
df_corr=df_corr[df_corr["pvalue"]<0.2]
df_corr.head()
````

````python
# Plot display the correlation

df_corr_pivot=df_corr.pivot_table(index="x",columns="y",values="corr").reset_index()

fig,axs=plt.subplots(1,4,figsize=(10,8),sharey=True,dpi=100)

sns.barplot(x="RevenueByRetail_Marketplace_Revenue",y="x",ax=axs[0],data=df_corr_pivot)
axs[0].set_ylabel("")
axs[0].set_xlabel("Retail_Marketplace_Revenue",fontsize=10,rotation=15)
sns.despine()

sns.barplot(x="Total_DeseasonalizedMarketplace_Revenue",y="x",ax=axs[1],data=df_corr_pivot)
axs[1].set_ylabel("")
axs[1].set_xlabel("DeseasonalizedMarketplace_Revenue",fontsize=10,rotation=15)
sns.despine()

sns.barplot(x="Total_Marketplace_SeasonalIndex",y="x",ax=axs[2],data=df_corr_pivot)
axs[2].set_ylabel("")
axs[2].set_xlabel("Marketplace_SeasonalIndex",fontsize=10,rotation=15)
sns.despine()

sns.barplot(x="Total_Total_Visits",y="x",ax=axs[3],data=df_corr_pivot)
axs[3].set_ylabel("")
axs[3].set_xlabel("Total_Visits",fontsize=10,rotation=15)
sns.despine()

plt.suptitle("Features Correlation with Four Targets",fontsize=14)
plt.tight_layout()
````

## The PCA Reduction of Several Macro Variables 

+ There are three variables that represent macroï¼ŒAccording to the variance interpretation diagram below, the first principal component alone explains all the variances of the three variables.So it is appropriate that we reduce the dimension of the three variables to one

````python
from sklearn.decomposition import PCA

pca_macro = PCA(random_state=1)
pca_macro.fit(df_model[['Macro_Total_Population','Macro_Total_RealDisposablePersonalIncomePerCapita','Macro_Total_TotalVehicleSales']])

exp_var = pca_macro.explained_variance_ratio_
df_pca = pd.DataFrame(enumerate(exp_var))
df_pca.columns = ['N', 'explained variance ratio']
df_pca["N"]=df_pca["N"]+1
df_pca['cumulative explained variance ratio'] = np.cumsum(df_pca['explained variance ratio'])

display(df_pca)
fig, axes = plt.subplots(1, 2, figsize = (12, 3))
_=df_pca[['N','explained variance ratio']].set_index('N').plot(kind="line",
                                          marker="o",
                                          ax = axes[0],
                                          title="explained variance ratio")
_=axes[0].set_xticks([1,2,3],minor=False)

_=df_pca[['N','cumulative explained variance ratio']].set_index('N').plot(kind="line",
                                                     marker="o",
                                                     ax = axes[1],
                                                     title="cumulative explained variance ratio")
_=axes[1].set_xticks([1,2,3],minor=False)
````

````python
# The PCA model with only 1 principal component was retrained

pca_macro = PCA(random_state=1,n_components=1)
pca_macro.fit(df_model[['Macro_Total_Population','Macro_Total_RealDisposablePersonalIncomePerCapita','Macro_Total_TotalVehicleSales']])

df_model["macro_pca"]=pca_macro.transform(df_model[['Macro_Total_Population','Macro_Total_RealDisposablePersonalIncomePerCapita','Macro_Total_TotalVehicleSales']])
````

## model: Total_DeseasonalizedMarketplace_Revenue

+ Since the data is not enough, we are not decide to take partition of train and test dataset 
+ Take the linear regression and see the coefficents of each variables

````python
from sklearn.linear_model import Lasso
from sklearn.preprocessing import StandardScaler
````

````python
# Standardizing the features

scale=StandardScaler()
df_model_scale=df_model.copy()

for col in ['BrandMedia_ConnectedTV_Spends', 'BrandMedia_Default_Spends','BrandMedia_Display_Spends', 'BrandMedia_OOH_Spends',
'BrandMedia_Radio_Spends', 'BrandMedia_Social_Spends','BrandMedia_TotalDigitalVideo_Spends', 'PerformanceMedia_App_Spends',
'PerformanceMedia_Display_Spends', 'PerformanceMedia_PaidSearch_Spends','PerformanceMedia_Partners_Spends','macro_pca', 'comp_pca']:
    
    df_model_scale[col]=scale.fit_transform(df_model_scale[[col]])

df_model_scale.head()
````

````python
model_dmr=Lasso(random_state=1)

X=df_model_scale[['BrandMedia_ConnectedTV_Spends', 'BrandMedia_Default_Spends','BrandMedia_Display_Spends', 'BrandMedia_OOH_Spends',
'BrandMedia_Radio_Spends', 'BrandMedia_Social_Spends','BrandMedia_TotalDigitalVideo_Spends', 'PerformanceMedia_App_Spends',
'PerformanceMedia_Display_Spends', 'PerformanceMedia_PaidSearch_Spends','PerformanceMedia_Partners_Spends','macro_pca', 'comp_pca']]
y=df_model_scale['Total_DeseasonalizedMarketplace_Revenue']

model_dmr.fit(X,y)

print("Model R2: {:0.4f}".format(model_dmr.score(X,y)))
````

````python
# The scatterplot shows the relationship between actual values and prediction

plt.figure(figsize=(8,5))

sns.regplot(y,model_dmr.predict(X))
sns.despine()
# plt.xscale("log")
# plt.yscale("log")
plt.xlabel("Total_DeseasonalizedMarketplace_Revenue",fontsize=12)
plt.ylabel("Predicted Value",fontsize=12)
plt.title("Scatter Plot of True vs Predicted Values",fontsize=14)
plt.grid(alpha=0.2)
````

````python
# The coefficients of each features

fig,ax=plt.subplots(1,1,figsize=(3,5),dpi=100)

df_model_coef_tmp_dmr=pd.DataFrame(zip(X.columns,model_dmr.coef_),columns=["features","coefficient"])
df_model_coef_tmp_dmr=df_model_coef_tmp_dmr.sort_values(by="coefficient",ascending=False)

sns.barplot(x="coefficient",y="features",ax=ax,data=df_model_coef_tmp_dmr)
axs[0].set_ylabel("")
axs[0].set_xlabel("Coefficient",fontsize=10,rotation=15)
sns.despine()
````

+ Analysis of coefficientï¼š
    - PerformanceMedia_Partners_Spendsï¼Œ BrandMedia_TotalDigitalVideo_Spendsï¼Œ BrandMedia_Display_Spends,etc. The spends of total six channel have the positive relationship with eseasonalizedMarketplace_Revenue
    - Take PerformanceMedia_Partners_Spends as exampleï¼ŒAfter standardization, every PerformanceMedia_Partners_Spends increase 1 unitï¼Œthe revenue would increase around 750 thousand ï¼ˆ75wï¼‰
    - It is not worthwhile to spend on PerformanceMedia_PaidSearch_Spends and some last channel, we suggest to decrease the spend in the future.
